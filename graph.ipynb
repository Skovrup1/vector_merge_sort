```python
import matplotlib.pyplot as plt
import numpy as np

array_sizes = [2 ** i for i in range(10, 17)]
cpp_times = [6, 8, 25, 35, 64, 146, 315]
ispc_times = [9, 28, 32, 72, 162, 356, 811]
c_times = [8, 17, 34, 73, 152, 309, 662]
ispc_sse4_times = [20, 58, 105, 234, 516, 1135, 2467]

cpp_throughput = [65.1042, 97.6562, 62.5000, 89.2857, 97.6562, 85.6164, 79.3651]
ispc_throughput = [43.4028, 27.9018, 48.8281, 43.4028, 38.5802, 35.1124, 30.8261]
c_throughput = [48.8281, 45.9559, 45.9559, 42.8082, 41.1184, 40.4531, 37.7644]
ispc_sse4_throughput = [19.5312, 13.4698, 14.8810, 13.3547, 12.1124, 10.1338, 10.1338]

plt.figure(figsize=(10, 5))
plt.plot(array_sizes, cpp_times, marker='o', label='CPP, sequential')
plt.plot(array_sizes, ispc_times, marker='o', label='ISPC (avx2), sequential')
plt.plot(array_sizes, c_times, marker='o', label='C, sequential')
plt.plot(array_sizes, ispc_sse4_times, marker='o', label='ISPC (sse4), sequential')
plt.xlabel('Array Size')
plt.ylabel('Compute Time (microseconds)')
plt.title('Compute Time vs Array Size (Linear Scale)')
plt.legend()
plt.grid(True, which="both", ls="--")
plt.xticks(array_sizes, [f'{size}' for size in array_sizes], rotation=45)
plt.savefig('compute_time_plot.png', dpi=300, bbox_inches='tight')
plt.show()

plt.figure(figsize=(10, 5))
plt.plot(array_sizes, cpp_throughput, marker='o', label='CPP, sequential')
plt.plot(array_sizes, ispc_throughput, marker='o', label='ISPC (avx2), sequential')
plt.plot(array_sizes, c_throughput, marker='o', label='C, sequential')
plt.plot(array_sizes, ispc_sse4_throughput, marker='o', label='ISPC (sse4), sequential')
plt.xlabel('Array Size')
plt.ylabel('Throughput (MiB/s)')
plt.title('Throughput vs Array Size (Linear Scale)')
plt.legend()
plt.grid(True, which="both", ls="--")
plt.xticks(array_sizes, [f'{size}' for size in array_sizes], rotation=45)
plt.savefig('throughput_plot.png', dpi=300, bbox_inches='tight')
plt.show()
```
